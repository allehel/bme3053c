```python
#Homework 1
#Lehel
def array_analysis(bp_readings):
  bp_max = max(bp_readings)
  bp_min = min(bp_readings)
  normalized_bps = []
  for bp in bp_readings:
    normalized_value = (bp - bp_min)/(bp_max - bp_min)
    normalized_bps.append(normalized_value)
  return normalized_bps

bp_readings = [120, 137, 151, 123, 162, 173, 127, 154]
new_bp_list = array_analysis(bp_readings)
print("Normalized bp list:", new_bp_list)



```
